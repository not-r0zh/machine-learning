```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

***

Gráfica de  <h2>$ x² $</h2>

```python
b = np.array([])
xd = np.linspace(-100,100,100000)
for i in xd:
    b = np.append(b, [i*i])
plt.plot(xd,b)
```

***

Grafica de <h2>$ sin(x) $</h2>

```python
b = np.array([])
xd = np.linspace(0,2*math.pi,10000)
for i in xd:
    b = np.append(b,[math.sin(i)])
plt.plot(xd,b)
```

Grafica de <h2>$ ln(x) $</h2>

```python
b = np.array([])
xd = np.linspace(0.00001,30,100000)
for i in xd:
    b = np.append(b,[math.log(i)])
plt.xlim([0,20])
plt.ylim([-3,3])
plt.plot(xd,b)
```

Grafica de <h2>Circulo XD</h2>

```python
radio = int(input("Introducza el radio\n"))
x = np.linspace(-radio,radio,10000)
y = np.array([])
for i in x:
    y = np.append(y,[math.sqrt(radio*radio-i*i)])
plt.subplot().set_aspect(1) 
plt.grid()
plt.xlabel("x")
plt.ylabel("y",rotation=0)
plt.plot(x,y)
plt.plot(-x,-y)
```
