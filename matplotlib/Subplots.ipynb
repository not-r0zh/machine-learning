```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2*np.pi*t)
s2 = np.sin(4*np.pi*t)

f = plt.figure()
f.set_figwidth(4)
f.set_figheight(1)

g = plt.figure()
g.set_figwidth(14)
g.set_figheight(1)

plt.figure(f)
plt.subplot(131)
plt.plot(t, s1)
plt.figure(f)
plt.subplot(132)
plt.plot(t, 2*s1)
plt.show()
plt.figure(g)
plt.subplot(133)
plt.plot(t, s2)
```

```python
fig, ax = plt.subplots(2, 2)
ax[0,0].plot(t, s1)
ax[0,1].plot(t, s2)
ax[1,0].plot(t, s2)
ax[1,1].plot(t, s1)
fig.tight_layout()
plt.show()
```
